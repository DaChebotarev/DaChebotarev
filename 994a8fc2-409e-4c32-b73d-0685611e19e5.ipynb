```python
#–∑–∞–º–µ–Ω–∏–º –ø—Ä–æ–ø—É—Å–∫–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ kitchen_area 
#—Ä–∞–∑–¥–µ–ª–∏–º –≤—Å–µ –æ–±—ä–µ–∫—Ç—ã –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ –Ω–∞ 4 –≥—Ä—É–ø–ø—ã –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –æ–±—â–µ–π –ø–ª–æ—â–∞–¥–∏ - total_area
def categorize_total_area(area):
    try:
        if 0 <= area <= 10:
            return 'XXXS'
        elif 10 <= area <= 20:
            return 'XXS'
        elif 20 <= area <= 30:
            return 'XS'
        elif 30 <= area <= 40:
            return 'S'
        elif 40 <= area <= 50:
            return 'M'
        elif 50 <= area <= 60:
            return 'L'
        elif 60 <= area <= 70:
            return 'XL'
        elif 70 <= area <= 80:
            return 'XXL'
        elif 80 <= area <= 90:
            return 'XXXL'
        elif area >= 90:
            return 'XXXXL'
    except:
        pass
```

```python
#–¥–æ–±–∞–≤–∏–º —Å—Ç–æ–ª–±–µ—Ü —Å —Ä–∞–∑–º–µ—Ä–æ–º –∫–≤–∞—Ä—Ç–∏—Ä –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
data['total_area_size'] = data['total_area'].apply(categorize_total_area)
```

```python
#–∑–∞–º–µ–Ω–∏–º –ø—Ä–æ–ø—É—Å–∫–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ living_area –Ω–∞ –º–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Ç—Å–∏ –æ—Ç —Ä–∞–Ω–µ–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞ –∫–≤–∞—Ä—Ç–∏—Ä—ã  
for i in data['total_area_size'].unique():
    data.loc[(data['living_area'].isna())& (data['total_area_size']==i), 'living_area'] = \
    data.loc[data['total_area_size']==i, 'living_area'].median()
```
